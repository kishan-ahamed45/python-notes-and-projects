## python libraries 

**Python libraries are collections of pre-written code that help you perform tasks easily without writing everything from scratch.**

### 1.Pandas

Pandas is a Python library used for easy data handling, analysis, and manipulation using tables called DataFrames.

```python
import pandas as pd
```

```python
insurance_data = pd.read_csv( filepath_or_buffer = "insurance.csv")
```

```python
insurance_data
```

### 2.NumPy 

NumPy is a Python library used for fast numerical calculations, arrays, and mathematical operations.

```python
import numpy as np
```

```python
np.repeat(a = insurance_data , repeats = 3 , axis= 1)
```

```python
insurance_data_updated =pd.DataFrame(data= np.repeat(a = insurance_data , repeats = 3 , axis= 1))
insurance_data_updated
```

### 3.Matplotlib , Seaborn , Plotly

Matplotlib: A Python library used to create basic charts and graphs like line, bar, and pie charts.

Seaborn: A Python library built on Matplotlib for making attractive and statistical data visualizations easily.

Plotly: A Python library used to create interactive and dynamic graphs that can be viewed in web browsers.

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(x = insurance_data.age)
```

```python
import seaborn as sns
```

```python
sns.histplot(x = insurance_data.age)
```
