## Pandas – Introduction

**Pandas is a Python library used for data manipulation and data analysis.**

**It helps us work with tabular data such as CSV files and Excel sheets easily and efficiently.**

**Pandas provides two main data structures:**

* Series – one-dimensional data

* DataFrame – two-dimensional data (rows and columns)

**Pandas is widely used in Data Science, Data Analysis, and Machine Learning.**
