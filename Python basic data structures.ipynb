## Python basic data structures

#### A data structure is a way to store and organize data in a computer so that we can access and modify it efficiently.

* In Python, data structures are built-in types that help us handle data easily.

**Common 4 Python Data Structures:**

* List

* Tuple

* Set

* Dictionary

### 1. List

**A list is an ordered collection of items in Python.**

* It is mutable → you can change, add, or remove items.

* It allows duplicates.

* Defined using square brackets []

* Lists are heterogeneous, meaning they can store multiple data types in a single list.

```python
empty_cart = []
```

```python
empty_cart
```

```python
type(empty_cart)
```

```python
shopping_list = ["Rice", "Milk", "Eggs", "Bread", "Sugar", "Tea" , "Rice" , 9.99 , 69]
shopping_list
```

```python
type(shopping_list)
```

```python
len(shopping_list)
```

```python
shopping_list[0]
```

```python
shopping_list[0:3]
```

```python
shopping_list[:3]
```

```python
shopping_list[2] = "fruits"
```

```python
shopping_list
```

### 2. Set

**Definition:**
* A set is an unordered collection of items in Python.

* It is mutable (you can add or remove items)

* It does NOT allow duplicate values

* Defined using curly braces {}

```python
empty_cart02 = set()
```

```python
empty_cart02 
```

```python
type(empty_cart02 )
```

```python
shopping_list02 = {"Rice", "Milk", "Eggs", "Bread", "Sugar", "Tea" , "Rice" , 9.99 , 69}
shopping_list02
```

```python
type(shopping_list02)
```

```python
shopping_list02 [0]
```

```python
shopping_list02 [2] = "fruits"
```

```python
shopping_list02.add("fruits")
```

```python
shopping_list02
```

```python
shopping_list02.remove("fruits")
shopping_list02
```

### Tuple

**Definition:**
* A tuple is an ordered collection of items in Python.

* It is immutable → values cannot be changed after creation

* It allows duplicate values

* Defined using parentheses () 

```python
empty_cart03 = ()
empty_cart03
```

```python
type(empty_cart03)
```

```python
shopping_list03 = ("Rice", "Milk", "Eggs", "Bread", "Sugar", "Tea" , "Rice" , 9.99 , 69)
shopping_list03
```

```python
shopping_list03 [0]
```

```python
shopping_list03[0:3]
```

```python
shopping_list03 [2] = "fruits"
```
